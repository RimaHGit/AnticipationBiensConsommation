```python
# split train & test data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, X_num_train, X_num_test, X_fact_train, X_fact_test, X_sur_train, X_sur_test, Y1_train, Y1_test, Y2_train, Y2_test = train_test_split(X, X_num, X_fact, X_sur, Y1, Y2, test_size = 0.3) # test data est 30% des données orginales
                                                                                                                                                                                                        # test_size = 0.3
```

```python
X_train.shape, X_num_train.shape, X_fact_train.shape, X_sur_train.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import mean_squared_error
```

## Target 1 : Log_TotalGHGEmissions(MetricTonsCO2e)

```python
# Test de différent model sur toutes les features
models = [LinearRegression(), KNeighborsRegressor(), DecisionTreeRegressor()]

for model in models:
    model.fit(X_train, Y1_train)
    predictions = model.predict(X_train)
    print("Train results")
    print(type(model).__name__, mean_squared_error(Y1_train, predictions))
    print("Test results")
    predictions = model.predict(X_test)
    print(type(model).__name__, mean_squared_error(Y1_test, predictions))
    print("-"*30)
```

```python
# Test de différent model sur toutes les features numériques uniquement (retrait des features categorielles)
models_num = [LinearRegression(), KNeighborsRegressor(), DecisionTreeRegressor()]

for model_num in models_num:
    model_num.fit(X_num_train, Y1_train)
    predictions = model_num.predict(X_num_train)
    print("Train results")
    print(type(model_num).__name__, mean_squared_error(Y1_train, predictions))
    print("Test results")
    predictions = model_num.predict(X_num_test)
    print(type(model_num).__name__, mean_squared_error(Y1_test, predictions))
    print("-"*30)
```

```python
# Test de différent parametres pour Decision tree
models = [DecisionTreeRegressor(max_depth=2), DecisionTreeRegressor(max_depth=10), DecisionTreeRegressor(max_depth=20), DecisionTreeRegressor(max_depth=30)]

for model in models:
    model.fit(X_train, Y1_train)
    predictions = model.predict(X_train)
    print("Train results")
    print(type(model).__name__, mean_squared_error(Y1_train, predictions))
    print("Test results")
    predictions = model.predict(X_test)
    print(type(model).__name__, mean_squared_error(Y1_test, predictions))
    print("-"*30)
```

```python
# Test de différent parametres pour KNN
n_neighbors_list = [2, 5, 10]
for n_neighbors in n_neighbors_list :
    model = KNeighborsRegressor(n_neighbors=n_neighbors)
    model.fit(X_train, Y1_train)
    predictions = model.predict(X_train)
    print(n_neighbors)
    print("Train results")
    print(type(model).__name__, mean_squared_error(Y1_train, predictions))
    print("Test results")
    predictions = model.predict(X_test)
    print(type(model).__name__, mean_squared_error(Y1_test, predictions))
    print("-"*30)
```

```python
# le meilleur modèle est celui avec la valeur la plus petite de mean_squared_error
model = models[1]
model
```

```python
predictions1 = model.predict(X_test)
df = X_test.copy()
df['Predictions_Log_TotalGHGEmissions(MetricTonsCO2e)'] = predictions1
df['Actual'] = Y1_test
df["Error"] = Y1_test - predictions1
df['RelativeError'] = 100*abs(Y1_test - predictions1)/Y1_test
avgerror = sum(df['RelativeError'])/len(df['RelativeError'])
avgerror
```
